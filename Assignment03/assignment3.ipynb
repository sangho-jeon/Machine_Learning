```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

path = "data.csv"
data = np.genfromtxt(path, delimiter=',')

x_data = data[:, 0]
y_data = data[:, 1]

m=x_data.size

plt.figure(figsize=(8, 8))
plt.scatter(x_data, y_data, c='black', alpha=0.3)
plt.show()
```

```python

def objective_function(y, hx, m):
    loss =  hx - y
    tmp = loss*loss
    sqr = np.sum(tmp) 
    return sqr/(2*m)

def obj_f(y, t0, t1, x, m):
    hx = t0+t1*x
    loss = hx-y
    tmp = loss*loss
    sqr = np.sum(tmp)
    return sqr/(2*m)

def loss_function_theta_zero(y, hx, m):
    loss = hx - y
    tmp = np.sum(loss)
    return tmp/m


    
def loss_function_theta_one(y,hx, x, m):
    loss = (hx - y)*x
    tmp = np.sum(loss)
    return tmp/m
    
```

```python
threshold = 30

#initial condition
theta0 = -30
theta1 = -30
hx = x_data*theta1 + theta0
#step_size 
step = 0.001
loss_list = []
loss_count = []
theta0_list = []
theta1_list = []
enough = False
count = 0 
while(enough == False):
    loss = objective_function(y_data, hx, m)
    loss_list.append(loss)
    loss_count.append(count)
    theta0_list.append(theta0)
    theta1_list.append(theta1)
    if(loss < threshold ):
        enough = True
    else:
        count = count + 1
        theta0 = theta0 - step*loss_function_theta_zero(y_data, hx, m)
        theta1 = theta1 - step*loss_function_theta_one(y_data, hx, x_data, m)
        hx = x_data*theta1 + theta0
        
print(theta0, theta1)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(x_data, y_data, c='black', alpha=0.3)
plt.plot([-10,10],[theta0-theta1*10, theta1*10+theta0], c='r')
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

energy_t0 = np.arange(-30, 30, 1)
energy_t1 = np.arange(-30, 30, 1)
t0, t1 = np.meshgrid(energy_t0, energy_t1)
energy_z = []

for i in energy_t1:
    line = []
    for j in energy_t0:
        line.append(obj_f(y_data, j, i, x_data, x_data.size))
    energy_z.append(line)
    

z=np.array(energy_z)
surf = ax.plot_surface(t0,t1,z,cmap='coolwarm',linewidth=0,antialiased=True)
plt.show()
```

```python

Loss = np.array(loss_list)
Theta0 = np.array(theta0_list)
Theta1 = np.array(theta1_list)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(t0,t1,z,cmap='coolwarm',linewidth=0,antialiased=True,alpha = 0.3)
ax.scatter(Theta0,Theta1,Loss, zdir=Loss, color="yellow")
plt.show()
```
