```python
import numpy as np
import matplotlib.pyplot as plt

data    = np.genfromtxt("data-nonlinear.txt", delimiter=',')
pointX  = data[:, 0]
pointY  = data[:, 1]
label   = data[:, 2]

x = np.array(pointX)
y = np.array(pointY)


pointX0 = pointX[label == 0]
pointY0 = pointY[label == 0]

pointX1 = pointX[label == 1]
pointY1 = pointY[label == 1]

x0 = np.array(pointX0)
y0 = np.array(pointY0)
x1 = np.array(pointX1)
y1 = np.array(pointY1)
```

```python
def z(x,y,theta):
    add = []
    for i in range(0,100):
        r = i/10
        r = int(r)
        c = i%10
        add.append(theta[i]*(x**r)*(y**c))
    total = np.array(add)
    result = total.sum(axis = 0)
    return result


def sigma(z):
    return 1/(1+np.exp(-1*z))

def objective(sigma, i):
    if i == 1:
        tmp = np.log(sigma)
        tmp = -1 * tmp
        return np.sum(tmp)
    elif i == 0:
        tmp = np.log(1 - sigma)
        tmp = -1*tmp
        return np.sum(tmp)

    
def objective_ramda(total, ramda, theta):
    return total + np.sum(theta*theta)*ramda/2




def score(z, i):
    count = 0
    for k in z:
        if abs(k-i) <= 0.5:
            count+=1
    return count
    

def resulting(x, y, theta):
    tmp = 0
    for i in range(0,100):
        r = int(i/10)
        c = i%10
        tmp += theta[i]*(x**r)*(y**c)
    result = 1/(1+np.exp(-1*tmp))
    return result


```

```python
theta_under = np.zeros(100)
theta_just = np.zeros(100)
theta_over = np.zeros(100)
enough = False
ramda_under = 0.1
ramda_just = 0.01
ramda_over = 0.001
step = 0.1
loss_under_iter = []
loss_just_iter = []
loss_over_iter = []
accuracy_under = []
accuracy_just = []
accuracy_over = []
loss_count = []
count = 0
m = x0.size + x1.size

while enough == False:
    loss_count.append(count)
    tmp0 = z(x0, y0, theta_under)
    tmp1 = z(x1, y1, theta_under)
    s0 = sigma(tmp0)
    s1 = sigma(tmp1)
    score0 = score(s0,0)
    score1 = score(s1,1)
    accuracy_under.append(score0+score1)
    t0 = objective(s0, 0)
    t1  = objective(s1, 1)
    total = (t0+t1)/(x0.size + x1.size)
    loss_under = objective_ramda(total, ramda_under, theta_under)

    tmp0 = z(x0, y0, theta_just)
    tmp1 = z(x1, y1, theta_just)
    s0 = sigma(tmp0)
    s1 = sigma(tmp1)
    score0 = score(s0,0)
    score1 = score(s1,1)
    accuracy_just.append(score0+score1)
    t0 = objective(s0, 0)
    t1  = objective(s1, 1)
    total = (t0+t1)/(x0.size + x1.size)
    loss_just = objective_ramda(total, ramda_just, theta_just)

    tmp0 = z(x0, y0, theta_over)
    tmp1 = z(x1, y1, theta_over)
    s0 = sigma(tmp0)
    s1 = sigma(tmp1)
    score0 = score(s0,0)
    score1 = score(s1,1)
    accuracy_over.append(score0+score1)
    t0 = objective(s0, 0)
    t1  = objective(s1, 1)
    total = (t0+t1)/(x0.size + x1.size)
    loss_over = objective_ramda(total, ramda_over, theta_over)

    loss_under_iter.append(loss_under)
    loss_just_iter.append(loss_just)
    loss_over_iter.append(loss_over)
    count += 1
    if(count == 6000):
        enough = True
    else:
        z0 = z(x0, y0, theta_under)
        z1 = z(x1, y1, theta_under)
        s0 = sigma(z0)
        s1 = sigma(z1)
        for i in range(0,100):
            r = int(i/10)
            c = i%10
            p0 = (s0)*(x0**r)*(y0**c) + ramda_under*(theta_under[i])
            p1 = (s1-1)*(x1**r)*(y1**c) + ramda_under*(theta_under[i])
            t0 = np.sum(p0) 
            t1 = np.sum(p1)
            theta_under[i] = theta_under[i] - step*((t0+t1)/(m)) 
            
        z0 = z(x0, y0, theta_just)
        z1 = z(x1, y1, theta_just)
        s0 = sigma(z0)
        s1 = sigma(z1)
        for i in range(0,100):
            r = int(i/10)
            c = i%10
            p0 = (s0)*(x0**r)*(y0**c) + ramda_just*(theta_just[i])
            p1 = (s1-1)*(x1**r)*(y1**c) + ramda_just*(theta_just[i])
            t0 = np.sum(p0) 
            t1 = np.sum(p1)
            theta_just[i] = theta_just[i] - step*((t0+t1)/(m)) 

        z0 = z(x0, y0, theta_over)
        z1 = z(x1, y1, theta_over)
        s0 = sigma(z0)
        s1 = sigma(z1)
        for i in range(0,100):
            r = int(i/10)
            c = i%10
            p0 = (s0)*(x0**r)*(y0**c) + ramda_over*(theta_over[i])
            p1 = (s1-1)*(x1**r)*(y1**c) + ramda_over*(theta_over[i])
            t0 = np.sum(p0) 
            t1 = np.sum(p1)
            theta_over[i] = theta_over[i] - step*((t0+t1)/(m)) 

    
        
accuracy_just = (np.array(accuracy_just)/118)*100
accuracy_under = (np.array(accuracy_under)/118)*100
accuracy_over = (np.array(accuracy_over)/118)*100



```

## Results

* ###  Plotting Training Data

```python
plt.figure()
plt.scatter(pointX0, pointY0, c='b')
plt.scatter(pointX1, pointY1, c='r')
plt.tight_layout()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

* ### Plotting Training Error & Displaying regularization parameters

```python
plt.figure(figsize=(8, 8))
plt.plot(loss_count, loss_under_iter,  color='b' )
plt.plot(loss_count, loss_just_iter,  color='g' )
plt.plot(loss_count, loss_over_iter,  color='r' )
plt.text(5000, 0.43, "overfitting ramda = 0.001", color='red')
plt.text(5000, 0.63, "underfitting ramda = 0.1", color='blue')
plt.text(5000, 0.52, "just-right-fitting ramda = 0.01", color='green')
plt.show()
```

* ### Plotting Training Accuracy & Displaying final training accuracy

```python
plt.figure(figsize=(6, 6))

plt.plot(loss_count, accuracy_under,  color='green')
plt.plot(loss_count, accuracy_just ,  color='red')
plt.plot(loss_count, accuracy_over,  color='blue')

finalunder = int(accuracy_under[-1])
finaljust = int(accuracy_just[-1])
finalover = int(accuracy_over[-1])


plt.text(5000, 85, "overfitting accuracy = %d%%"%finalover , color='red')
plt.text(5000, 73, "underfitting accuracy = %d%%"%finalunder, color='green')
plt.text(5000, 80, "just-right-fitting accuracy = %d%%"%finaljust, color='blue')

plt.show()
```

* ### Plotting optimal classifier with varying regularization parameters superimposed on the training data

```python
tx = np.arange(-1.2, 1.5, 0.005)
ty = np.arange(-1.2, 1.5, 0.005)

zet_under = []
zet_just = []
zet_over = []
for i in ty:
    line0 = []
    line1 = []
    line2= []
    for j in tx:
        result_under = resulting(j,i,theta_under)
        result_just = resulting(j,i,theta_just)
        result_over = resulting(j,i,theta_over)
        if result_under>0.5:
            result_under = 1
        elif result_under <0.5:
            result_under = 0
        if result_just>0.5:
            result_just = 1
        elif result_just <0.5:
            result_just = 0
        if result_over>0.5:
            result_over = 1
        elif result_over <0.5:
            result_over = 0
        line0.append(result_under)
        line1.append(result_just)
        line2.append(result_over)
    zet_under.append(line0)
    zet_just.append(line1)
    zet_over.append(line2)

tmpx, tmpy = np.meshgrid(tx,ty)    
tz0 = np.array(zet_under)
tz1 = np.array(zet_just)
tz2 = np.array(zet_over)
plt.figure(figsize=(8, 8))
plt.contour(tmpx, tmpy, tz0)
plt.contour(tmpx, tmpy, tz1)
plt.contour(tmpx, tmpy, tz2)

plt.scatter(pointX0, pointY0, c='b')
plt.scatter(pointX1, pointY1, c='r')

plt.show()
```
